```python
# list all the runs in `https://wandb.ai/generating-faster/hf-bench`
import wandb

api = wandb.Api()
runs = api.runs("generating-faster/hf-bench")

# Print basic info for each run
for run in runs:
    print(f"Run {run.name}: {run.state}")
```

```python
# for each run, get its number of gpus and the type of gpus
for run in runs:
    hardware = "Unknown"
    if run.metadata:
        gpu_count = run.metadata.get('gpu_count', 'unknown')
        gpu_type = run.metadata.get('gpu', 'unknown')
        hardware = f"{gpu_count} * {gpu_type}"
    print(f"Run {run.name}: {gpu_count} * {gpu_type}")
```
